```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('portfolio_data.csv')

df['Date'] = pd.to_datetime(df['Date'])

df.set_index('Date', inplace=True)

selected_stocks = ['AMZN', 'NFLX']

for stock in selected_stocks:
    df[f'{stock}_Daily_Return'] = df[stock].pct_change()

plt.figure(figsize=(10, 6))
plt.scatter(df[f'{selected_stocks[0]}_Daily_Return'], df[f'{selected_stocks[1]}_Daily_Return'])
plt.title('Comparing')
plt.xlabel(f'{selected_stocks[0]} Daily Return')
plt.ylabel(f'{selected_stocks[1]} Daily Return')
plt.grid(True)
plt.show()

correlation = df[f'{selected_stocks[0]}_Daily_Return'].corr(df[f'{selected_stocks[1]}_Daily_Return'])
print(f'coefficient corr: {correlation}')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('portfolio_data.csv')
df['Date'] = pd.to_datetime(df['Date'])

df.set_index('Date', inplace=True)

selected_stocks = ['AMZN', 'NFLX']

for stock in selected_stocks:
    df[f'{stock}_Daily_Return'] = df[stock].pct_change()

correlation_matrix = df[[f'{selected_stocks[0]}_Daily_Return', f'{selected_stocks[1]}_Daily_Return']].corr()

import seaborn as sns

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Pairwise Corr')
plt.show()

print(correlation_matrix)
```

A reading of 44 percent indicates that the overall price movements of AMZN and NFLX tend to be consistent to some extent. However, this is not a very strong correlation.
