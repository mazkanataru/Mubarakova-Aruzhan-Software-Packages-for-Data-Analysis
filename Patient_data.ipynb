```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
file_path = 'Death_rates.csv'

df = pd.read_csv(file_path)
df.head(5)
```

```python
df.columns
```

```python


df.describe()

#"ESTIMATE"
df['ESTIMATE'].hist()
plt.title('Distribution of ESTIMATE')
plt.xlabel('ESTIMATE')
plt.ylabel('Frequency')
plt.show()

sns.lineplot(x='YEAR', y='ESTIMATE', data=df)
plt.title('Trends over Time')
plt.show()

# Correlation matrix
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
data = {
    'YEAR': [2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020],
    'ESTIMATE': [38, 40, 38, 45, 42, 50, 48, 55, 58, 60, 62, 65, 70]
}

df = pd.DataFrame(data)

# Filter data for the period between 2000 and 2010filtered_df = df[(df['YEAR'] >= 2010) & (df['YEAR'] <= 2020)]

# Data visualization
plt.figure(figsize=(10, 6))
sns.barplot(x='YEAR', y='ESTIMATE', data=df, color='skyblue')
plt.title('Suicide Rates between 2010 and 2020')
plt.xlabel('Year')
plt.ylabel('Suicide Rate')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    'AGE': [1, 2, 3, 4, 5, 6],
    'ESTIMATE': [5, 8, 10, 15, 18, 20]
}
df = pd.DataFrame(data)

# Data visualization using scatter plot
plt.figure(figsize=(10, 6))
sns.barplot(x='AGE', y='ESTIMATE', data=df)
plt.title('Dependency of Suicide Rate on Age')
plt.xlabel('Age (AGE)')
plt.ylabel('Suicide Rate (ESTIMATE)')
plt.show()
```
